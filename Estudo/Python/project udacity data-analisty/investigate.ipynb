```python
data_gun= pd.read_excel('gun_data.xlsx')
data_gun.head()
```

```python
# Load your data and print out a few lines. Perform operations to inspect data
#   types and look for instances of missing or possibly errant data.
data_census = pd.read_csv('U.S. Census Data.csv')
data_census.head(5)
```

```python
data_gun.info()
```

```python
#  lets fill the null values for the mean in both dataset 
```

```python
data_gun.fillna(data_gun.mean(),axis=0,inplace=True)
#fist sum is for each column the second is for the hole dataset
data_gun.isnull().sum().sum()
```

```python
#this code is for drop all nun values 
data_census=data_census.drop(index=data_census.iloc[65:].index,axis=1)
```

```python
#i drop fact note because we has just nan volues and this dont help in analisty
data_census=data_census.drop('Fact Note',axis=1)
```

```python
data_census.fillna(data_census.mean(),axis=0,inplace=True)
```

```python
data_census.isnull().sum().sum()
```

```python
#now he dont have problems with null values i see some values in data_census dont was in right datatype so lets fix this
```

```python
#fist in data_census;covert by string to float  r'D' is a regular expression
```

```python
states=[x for x in data_census.columns[2:]]
for state in states:
    data_census[state].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
    data_census[state]=pd.to_numeric(data_census[state], downcast='float', errors='ignore')

```

```python
# change the columns  of data_gun for the rigth datatype
col= list(data_gun.columns)
col = col[2:]
data_gun[col] = data_gun[col].applymap(num.int64)
data_gun.info()


```
